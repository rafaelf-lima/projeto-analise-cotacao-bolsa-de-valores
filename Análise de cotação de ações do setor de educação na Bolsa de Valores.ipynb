```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
```

```python
cotacao = web.DataReader('YDUQ3.SA', data_source = 'yahoo', start = '07-20-2021', end = '07-20-2022')
display(cotacao)
```

```python
cotacao['Adj Close'].plot(figsize = (15,10))
plt.title("Fechamento ajustado durante 20/07/2021 e 20/07/2022 - YDUQ3.SA", 
        fontsize='large',
        fontweight='bold',
        style='italic',
        family='monospace')
plt.show()
```
